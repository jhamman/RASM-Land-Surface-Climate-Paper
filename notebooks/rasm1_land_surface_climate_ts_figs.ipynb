```python
# Make the plot
def plot_by_mask(datasets, masks, variables, ylabels=None, figsize=(11, 8.5), colors=None, diff_key=None, sum_func=npb.nansum, wateryear=False):
    """
    datasets : OrderedDict of xray.Datasets
        dict of datasets with 
    masks : OrderedDict of numpy.NDArrays
        dict of 2d masks, keys in dictionary are used as titles
    rows : list like
        list of variable names to include in plot
    ylabels : list like, same size as rows
        list of ylabels (variable names) for plot
    figsize : tuple
        figure size, passed to pyplot.subplots
    """
    
    if ylabels is None:
        ylabels = variables
    
    fig, axes = plt.subplots(nrows=len(variables), ncols=len(masks), figsize=figsize, sharex=True, sharey='row')

    for i, (title, mask) in enumerate(masks.items()):

        ys, xs = np.nonzero(mask)
        areas = ncdata['rasm_domain']['area'].values[ys, xs]
        weights = (mask[ys, xs] * areas)
        weights /= weights.sum()

        for j, var in enumerate(variables):
            ax = axes[j, i]
            plt.sca(ax)
            ax.set_xlim((1, 12))
            

            if diff_key is not None:
                compare = sum_func(np.squeeze(datasets[diff_key][var].values)[:, ys, xs] * weights, axis=1)
            
            for name, ds in datasets.items():
                if var in ds:
                    
                    data = sum_func(np.squeeze(ds[var].values)[:, ys, xs] * weights, axis=1)
                    if diff_key is not None:
                        data -= compare
                    
                    if var == 'calc_albedo':
                        data[0] = np.nan
                        data[11:] = np.nan
                        print('Warning, Albedo cludge still in place')
                    
                    ax.plot(months, data, color=colors[name])

            if i == 0:
                ax.set_ylabel(ylabels[j])
            if j == 0:
                ax.set_title(title)

    xticks = np.arange(1, 13)
    ticks = [ax.set_xticks(xticks) for ax in axes[-1]]
    if wateryear:
        tick_labels = [ax.set_xticklabels(['S', 'O', 'N', 'D', 'J', 'F', 'M', 'A', 'M', 'J', 'J', 'A']) for ax in axes[-1]]
    else:
        tick_labels = [ax.set_xticklabels(['J', 'F', 'M', 'A', 'M', 'J', 'J', 'A', 'S', 'O', 'N', 'D']) for ax in axes[-1]]
    fig.tight_layout()
    fig.subplots_adjust(bottom=0.15)

    # add legend
    lines = []
    labels = []
    for source, color in colors.items():
        lines.append(mlines.Line2D([], [], color=color, label=var, linewidth=3))
        labels.append(source)
  
    leg = fig.legend(handles=lines, labels=labels, ncol=6, columnspacing=6,
                     fontsize=10, bbox_to_anchor=[0.52, 0.08], loc='upper center',
                     borderaxespad=0.05)
    leg.draw_frame(True)
    leg.get_frame().set_edgecolor('k')
    leg.get_frame().set_linewidth(0.5)
    
    return fig, axes
    
#     plt.tight_layout(h_pad=0, w_pad=0)
```

```python
def get_wy_months(ds, offset=4):
    months = ds['time.month']

    months += offset
    months[months > 12] -= 12
    return months

w_datasets = OrderedDict()
w_datasets['ERA'] = ncdata['era_monthly_ts1'][['Swq', 'calc_albedo']].sel(time=slice(start, end)).resample('MS', dim='time', how='mean')
w_datasets['ERA']['Tair'] = ncdata['era_monthly_ts2']['Tair'].sel(time=slice(start, end)).resample('MS', dim='time', how='mean')
w_datasets['ERA']['Precipitation'] = ncdata['era_monthly_ts3']['Precipitation'].sel(time=slice(start, end)).resample('MS', dim='time', how='mean')
w_datasets['ERA']['runoff_tot'] = ncdata['era_monthly_ts3']['runoff_tot'].sel(time=slice(start, end)).resample('MS', dim='time', how='mean')
w_datasets['ERA']['Evap'] = -1 * ncdata['era_monthly_ts3']['Evap'].sel(time=slice(start, end)).resample('MS', dim='time', how='mean')
w_datasets['MERRA'] = ncdata['merra_monthly_ts1'][['Precipitation', 'runoff_tot', 'Evap', 'Swq']].sel(time=slice(start, end)).resample('MS', dim='time', how='mean')
w_datasets['MERRA']['Tair'] = ncdata['merra_monthly_ts2']['Tair'].sel(time=slice(start, end)).resample('MS', dim='time', how='mean')
w_datasets['A2006'] = ncdata['adam_monthly_ts'][['Precipitation']].sel(time=slice(start, end)).resample('MS', dim='time', how='mean')

w_datasets['RASM$_{ERA}$'] = ncdata['rasm_era_monthly_ts'][['Tair', 'Precipitation', 'runoff_tot', 'Evap', 'Swq']].sel(time=slice(start, end)).resample('MS', dim='time', how='mean')
# datasets['$RASM_{CFSR}$'] = ncdata['rasm_cfsr_monthly_ts'][['Tair', 'Precipitation', 'runoff_tot', 'Evap', 'Swq']].sel(time=slice(start, end)).resample('MS', dim='time', how='mean')
w_datasets['VIC$_{S2006}$'] = ncdata['rasm_offline_monthly_ts'][['Tair', 'Precipitation', 'runoff_tot', 'Evap', 'Swq']].sel(time=slice(start, end)).resample('MS', dim='time', how='mean')

for k, ds in w_datasets.items():
    months = get_wy_months(ds)
    w_datasets[k] = ds.groupby(months).mean(dim='time')
```

```python
def calc_accum_prec(dsets):
    for k, v in dsets.items():
        if 'Precipitation' in v:
            print(k)
            d = v['Precipitation']
            days = xr.DataArray(np.roll(dpm['noleap'][1:], 4), dims=d['month'].dims, coords=d['month'].coords)
            dsets[k]['accum_prec'] = np.cumsum(d * days, axis=d.get_axis_num('month'))
    return dsets

w_datasets = calc_accum_prec(w_datasets)
```
